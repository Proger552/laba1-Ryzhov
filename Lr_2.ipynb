Задание 2. 

2.1.Преобразуйте полноцветное изображение в полутоновое в переменной Х.

2.2. Измените его яркость, добавив 10, 128 и 255. 

2.3. Оцените полученные результаты с помощью команды imshow.


```python
X = image.convert('L')
enchan = ImageEnhance.Brightness(X)

plt.subplot(1,3,1)
im_output_1 = enchan.enhance(1.1) #+10 
plt.imshow(im_output_1)
plt.subplot(1,3,2)
im_output_2 = enchan.enhance(2.28)#+128
plt.imshow(im_output_2)
plt.subplot(1,3,3)
im_output_3 = enchan.enhance(3.55)#+255
plt.imshow(im_output_3)
plt.show()
```
